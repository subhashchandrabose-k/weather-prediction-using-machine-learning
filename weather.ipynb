```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
weather=pd.read_csv("weatherAUS.csv")
weather
```

```python
weather.corr()
```

```python
weather['RainToday'].replace({'No':0,'Yes':1}, inplace=True)
weather['RainTomorrow'].replace({'No':0,'Yes':1}, inplace=True)
```

```python
weather.corr()
```

```python
weather.apply(lambda x: sum(x.isnull()),axis=0)
```

```python
weather1 = weather[pd.notnull(weather['RainToday'])]
```

```python
plt.scatter(weather1['RainTomorrow'],weather1['Rainfall']),weather1.boxplot(column='Rainfall',by='RainTomorrow')
```

```python
weather1["Pressure3pm"].value_counts()
```

```python
weather1["Pressure3pm"].fillna(1015.500,inplace=True)
weather1["Humidity3pm"].fillna(weather1["Humidity3pm"].mean(),inplace=True)
weather1["Temp3pm"].fillna(weather1["Temp3pm"].median(),inplace=True)
weather1["Cloud3pm"].fillna(weather1["Cloud3pm"].median(),inplace=True)
weather1["WindGustSpeed"].fillna(weather1["WindGustSpeed"].median(),inplace=True)
weather1["Sunshine"].fillna(weather1["Sunshine"].median(),inplace=True)
```

```python
weather1.apply(lambda x: sum(x.isnull()),axis=0)
```

```python
X=weather1[['Rainfall','Sunshine','WindGustSpeed','Humidity3pm','Pressure3pm','Cloud3pm','Temp3pm','RainToday']]
Y=weather1['RainTomorrow'] 
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.4,random_state=1)
```

```python
from sklearn.linear_model.logistic import LogisticRegression
lqsr=LogisticRegression(random_state=0)
lqsr.fit(X_train,Y_train) 
```

```python
X_test=X_test.values.reshape(-1,8)
```

```python
y_pred=lqsr.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
acc_sco=accuracy_score(Y_test,y_pred)
acc_sco
```
